```python
df1.drop(0,inplace=True)
df1.drop(1,inplace=True)
df1.head()
```

```python
df1=pd.melt(df1,id_vars=["Order_ID"],value_vars=["Consumer","Consumer","Consumer","Corporate","Corporate","Corporate",
                                                "Home Office","Home Office","Home Office"],var_name="Segment",value_name="Sales")
```
